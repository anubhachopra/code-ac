```python
import pandas as pd
```

```python
directory = 'C:/Users/ac4296/OneDrive - Drexel University/Anubha & Dr. T - Shared/Urban Regulatory projects/Urban Final'
file_path = f"{directory}/Data/legal_compendium_oct_2019_update_.xlsx"
output_path = f"{directory}/Output/Formating Urban"
```

```python
data= pd.read_excel(file_path, sheet_name='Front', header=1, engine='openpyxl')
```

```python
data.rename(columns={data.columns[0]: 'STATE'}, inplace=True)
data.tail()
```

```python
data= data.dropna(subset= ['STATE'])
```

```python
data[data['STATE'].isna()]
```

```python
data['STATE'].unique()
```

```python
final_df= pd.DataFrame()
for i in data['STATE'].unique():
    data_i= data[data['STATE']==i]
    df_transform= data_i.T.reset_index().rename(columns={'index':'Regulatory Type'})
    df_transform.rename(columns={df_transform.columns[1]: "Notes" }, inplace=True)
#     print(i)
#     print(data_i.shape)
#     print(df_transform.shape)
#     print(df_transform.head())
    df_transform.insert(0, 'STATE', i)
    final_df= pd.concat([final_df, df_transform])
```

```python
final_df.rename(columns={'Notes': 'Regulation Indicator'}, inplace=True)

final_df.shape
final_df.tail(50)
```

```python
final_df.to_csv(f'{output_path}/Step 1b Back(sources) transposed - Regulation Indicator.csv')
```
